## Imbalanced Dataset Handling

1. Upsampling 
2. Down Sampling

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)


## Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=5, scale=2, size=n_class_0),
    'target': [0] * n_class_0   
})


class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=3, scale=1, size=n_class_1),
    'feature2': np.random.normal(loc=8, scale=2, size=n_class_1),
    'target': [1] * n_class_1       
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

### UPSAMPLING

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
df_minority.head()
```

```python
df_majority.head()
```

```python
## Upsampling performed here
from sklearn.utils import resample
```

```python
df_minority_upsample = resample(df_minority,
                                 replace=True,
                                 n_samples=len(df_majority),    
                                 random_state=42
                                )
                                
```

```python
df_minority_upsample.value_counts()
```

```python
df_minority_upsample['target'].value_counts()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsample])
```

```python
df_upsampled['target'].value_counts()
```

```python
df_upsampled.head()
```

```python
df_upsampled.shape
```

## Downsampling (Jo zaida hai usko kam karna)

```python
class_0 = pd.DataFrame({
    'feature1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature2': np.random.normal(loc=5, scale=2, size=n_class_0),
    'target': [0] * n_class_0   
})


class_1 = pd.DataFrame({
    'feature1': np.random.normal(loc=3, scale=1, size=n_class_1),
    'feature2': np.random.normal(loc=8, scale=2, size=n_class_1),
    'target': [1] * n_class_1       
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
```

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
df_majority_downsample = resample(df_majority,
                                 replace=False,
                                 n_samples=len(df_minority
                                 ),    
                                 random_state=42
                                )
                                
```

```python
df_majority_downsample.shape
```

```python
pd.concat([df_minority, df_majority_downsample]).target.value_counts()
```
