# Normalization-Min Max Scaler

- Min-Max Normalization (or Min-Max Scaling) is used in machine learning to transform features to a common scale, typically within the range of 0 to 1.






```python
import seaborn as sns
```

```python
df=sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max=MinMaxScaler()
```

```python
min_max.fit(df[['distance','fare','tip']])
```

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.fit_transform(df[['distance','fare','tip']])
```

```python
min_max.transform([[1,3,4]])
```
