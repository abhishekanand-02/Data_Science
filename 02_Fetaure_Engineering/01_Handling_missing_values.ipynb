```python
df.dropna()  # Drops all rows with any missing values
```

```python
df.shape   # Original shape before dropping missing values
```

## Handling missing vlue by deleting column

```python
df.dropna(axis=1)  # Drops all columns with any missing values
```

## Imputation Techniques

1. Mean Value Imputation

```python
sns.displot(df['age'])  
```

```python
df.age.isnull().sum()
```

```python
df['Age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['Age_mean', 'age']]
```

2. Median Value Imputation

```python
df['Age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['Age_median','Age_mean', 'age']]
```

3. Mode Value Imputation ---> It is used mostly for categorical fetaures

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['age'].notna()]
```

```python
df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode = df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode
```

```python
df['embarked_mode'] = df['embarked'].fillna(mode)
```

```python
df[['embarked', 'embarked_mode']]
```

```python
df['embarked_mode'].isnull().sum()
```
