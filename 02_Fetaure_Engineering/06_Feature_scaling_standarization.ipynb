## Standardization --> We apply z-score = xi - x(mean)/ std

- We will get mean = 0 and standard deviation as 1
- Standardization is performed in machine learning to rescale features to have a mean of \(0\) and a standard deviation of \(1\), which improves model performance by ensuring all features are on a comparable scale


```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill=list(df['total_bill'])
```

```python
total_bill
```

```python
sns.histplot(total_bill)
```

```python
import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
mean,std
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std ## z_score formula
    normalized_data.append(z_score)

print(normalized_data)
```

```python
sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
import pandas as pd
pd.DataFrame(scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tips'])
```
