## Explaining Non-Linearity

```python
import numpy as np
import matplotlib.pyplot as plt


# Define the Linear Combination
w = [1,2] #weights
b= 3 # bias

# Create Data
x = np.linspace(-5,5,100)
y = w[0]*x + w[1]*x + b

# Plot the Linear Combination
plt.plot(x, y, label='y = w1*x + w2*x + b')
plt.title('Linear Combination')


plt.legend()

plt.show()


# Plot sigmoid activation
sigmoid = lambda x: 1 / (1 + np.exp(-x))
y_sigmoid = sigmoid(y)
plt.plot(x, y_sigmoid, label='Sigmoid Activation')
plt.title('Sigmoid Activation')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-10, 10, 100)

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Apply ReLU
y_relu = relu(x)

# Plot
plt.plot(x, y_relu, label='ReLU Activation', color='green')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input values
x = np.linspace(-10, 10, 100)

# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Apply Leaky ReLU
y_leaky = leaky_relu(x)

# Plot
plt.plot(x, y_leaky, label='Leaky ReLU Activation', color='orange')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh function
def tanh(x):
    return np.tanh(x)

# Input values
x = np.linspace(-10, 10, 100)
y = tanh(x)

# Plot
plt.plot(x, y, color='red')
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Input vector (logits)
z = np.array([2.0, 1.0, 0.1])

# Softmax function
def softmax(z):
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z)

# Apply Softmax
probabilities = softmax(z)

# Plot
plt.bar(range(len(z)), probabilities, color='purple')
plt.xticks(range(len(z)), ['Class 1', 'Class 2', 'Class 3'])
plt.title('Softmax Activation Function')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()
```
