## Measure of Dispersion

```python
ages_list = [23,23,43,56,32,68,98,45,32]
```

```python
import numpy as np
```

```python
np.mean(ages_list)
```

```python
## variance 
np.var(ages_list)
```

```python
## standard deviation
np.std(ages_list)
```

```python
import seaborn as sns
sns.histplot(ages_list, kde=True)
```
